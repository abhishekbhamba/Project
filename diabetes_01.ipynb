# (vii) Calculate the skewness value and divide variables into symmetrical, moderately skewed and highly skewed:-

```python
df1
```

```python
# here is the skewness of all columns with variable into moderately, symmetrically, highly skewed.
from scipy.stats import skew

for col in df1:
    print(col)
    print(skew(df1[col]))
    
    plt.figure()
    sns.distplot(df1[col])
    plt.show()
```
